# Univariate Analysis 

```python
sns.histplot(df["BMI"],kde=True,bins=30,color="Red")
plt.title("BMI distribution")
plt.show()
```

```python
sns.histplot(df["Age"],bins=30)
plt.title("Age Distribution")
plt.show()
```

```python
sns.countplot(data=df,x="HeartDiseaseorAttack")
plt.title("HeartDiseaseorAttack")
```

```python
sns.countplot(data=df,x="GenHlth",color="Red")
plt.title("General Health Distribution")
plt.show()
```

# Bivariate Analysis 

```python
sns.countplot(data=df, x='Smoker', hue='HeartDiseaseorAttack')
plt.show()
```

```python
sns.boxplot(data=df,x="HeartDiseaseorAttack",y="BMI")
plt.title("Heart Disease vs BMI")
plt.show()
```

```python
df.columns=df.columns.str.strip()
corr=df[["HeartDiseaseorAttack","HighBP","Smoker","Age"]]
sns.heatmap(corr.corr(),annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

```python
sns.scatterplot(data=df,x="BMI",y="Age",hue="HeartDiseaseorAttack")
plt.title("BMI vs Age")
plt.show()
```
